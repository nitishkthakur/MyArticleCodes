```python
def objective(x):
    return x**2


x = np.arange(-10, 10, .1)

y = objective(x)


plt.figure(figsize = (10, 5))
plt.plot(x, y)
plt.grid()
plt.ylabel('Cost', weight = 'bold')
plt.xlabel('Variable', weight = 'bold')
```

```python
def objective(x):
    return x**2 + (2*x**2 * np.sin(x)**2)


x = np.arange(-10, 10, .1)

y = objective(x)


plt.figure(figsize = (10, 5))
plt.plot(x, y)
plt.grid()
plt.ylabel('Cost', weight = 'bold')
plt.xlabel('Variable', weight = 'bold')
```

```python
%%cython -a
import numpy as np

def mandelbrot_cython(m, size, iterations):
    for i in range(size):
        for j in range(size):
            c = -2 + 3./size*j + 1j*(1.5-3./size*i)
            z = 0
            for n in range(iterations):
                if np.abs(z) <= 10:
                    z = z*z + c
                    m[i, j] = n
                else:
                    break
```
